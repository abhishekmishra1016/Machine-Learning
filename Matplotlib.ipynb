```python
#Subplot option 1
fig, ((ax1,ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter.keys(), nut_butter.values())
ax4.hist(np.random.randn(1000));
```

```python
#Subplot option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter.keys(), nut_butter.values())
ax[1, 1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv("heart_disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=100);
```

```python
over50 = heart_disease[heart_disease["age"]>50]
over50.head()
```

```python
#Pyplot method
over50.plot(kind='scatter', x='age', y='chol', c='target');
```

```python
#OO Method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
```

```python
#OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over50["age"], y=over50["chol"], c=over50["target"]);

#customize plot
ax.set(title="Heart Disease and Cholestrol level", xlabel="Age", ylabel="Cholestrol");

#add legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over50["chol"].mean(), linestyle='--');
```

```python
#subplots of chol, age , thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True);

#add data to ax0
scatter = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"]);

#customize plot
ax0.set(title="Heart Disease and Cholestrol level", ylabel="Cholestrol");

#add legend
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(over50["chol"].mean(), linestyle='--');

#add data to ax1
scatter = ax1.scatter(x=over50["age"], y=over50["thalach"], c=over50["target"]);

#customize plot
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");

#add legend
ax1.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax1.axhline(over50["thalach"].mean(), linestyle='--');


#add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
#Changing Style
plt.style.use('seaborn-whitegrid')

#OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over50["age"], y=over50["chol"], c=over50["target"], cmap="winter");    #cmap will change colour

#customize plot
ax.set(title="Heart Disease and Cholestrol level", xlabel="Age", ylabel="Cholestrol");

#add legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over50["chol"].mean(), linestyle='--');
```

```python
# Customizing y and x axis limitations
#subplots of chol, age , thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True);

#add data to ax0
scatter = ax0.scatter(x=over50["age"], y=over50["chol"], c=over50["target"], cmap="winter");

#customize plot
ax0.set(title="Heart Disease and Cholestrol level", ylabel="Cholestrol");

#setting x limit
ax0.set_xlim([50,80])

#add legend
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(over50["chol"].mean(), linestyle='--');

#add data to ax1
scatter = ax1.scatter(x=over50["age"], y=over50["thalach"], c=over50["target"], cmap="winter");

#customize plot
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#add legend
ax1.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax1.axhline(over50["thalach"].mean(), linestyle='--');


#add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');
```
